```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from sklearn.preprocessing import KBinsDiscretizer
import pandas as pd


pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
pd.options.display.max_colwidth = 199
pd.options.display.width = None
```

```python
%run ../util/load_data.py
%run ../util/data_visualization.py
%run ../util/outliers_statistics.py

data = load_data("../../Data")
train_data = data['train_data'][data['numerical_attributes'] + data['categorical_attributes'] + data['target']]
```

```python
discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='uniform')
discretizer.fit(train_data[data['numerical_attributes']])
discretized_values = discretizer.transform(train_data[data['numerical_attributes']])
```

```python
for idx, column in enumerate(data['numerical_attributes']):
    train_data[column] = [column + "[" + str(discretizer.bin_edges_[idx][int(x)]) + "-" + str(discretizer.bin_edges_[idx][int(x) + 1]) + "]" for x in discretized_values[:,idx]]
    
for column in data['categorical_attributes'] + data['target']:
    train_data[column] = train_data[column].apply(lambda x: column + "_" + str(x))
```

```python
def get_association_rules(dataset):
    te = TransactionEncoder()
    te_ary = te.fit(dataset).transform(dataset)
    df = pd.DataFrame(te_ary, columns=te.columns_)
    frequent_itemsets = apriori(df, min_support=0.4, use_colnames=True)

    print(frequent_itemsets)
    
    rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
    return rules
```

```python
transaction_dataset = train_data[data['numerical_attributes'] + data['categorical_attributes']].values
rules = get_association_rules(transaction_dataset)
```
